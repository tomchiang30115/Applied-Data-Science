# Data post-processing
## Converting Excel spreadsheet to csv

```python
import pandas as pd
import numpy as np
from datetime import *

# Read and store content
# of an excel file 
read_file = pd.read_excel("completed_form.xlsx")

read_file.dropna(subset = ['PrimaryActivityCode'], inplace=True)

read_file = read_file.drop(columns=['15mintime', '15min', 'weeks'])

read_file["StartDate"] = read_file["StartDate"].dt.strftime("%m/%d/%y")

read_file["StartTime"] = read_file["StartTime"].astype(str).str.replace(r'.{3}$', '', regex=True)
read_file["EndTime"] = read_file["EndTime"].astype(str).str.replace(r'.{3}$', '', regex=True)

read_file["UniBathWeekNo"] = read_file["UniBathWeekNo"].astype(str).str.replace(r'.{2}$', '', regex=True)
read_file["DurationMins"] = read_file["DurationMins"].astype(str).str.replace(r'.{2}$', '', regex=True)
read_file["EnjoymentScore"] = read_file["EnjoymentScore"].astype(str).str.replace(r'.{2}$', '', regex=True)

read_file["UniBathWeekNo"] = read_file["UniBathWeekNo"].astype(int)
read_file["DurationMins"] = read_file["DurationMins"].astype(int)
read_file["EnjoymentScore"] = read_file["EnjoymentScore"].astype(int)

# Write the dataframe object
# into csv file
read_file.to_csv("completed_data.csv", index = None, header=True)
    
# read csv file and convert 
# into a dataframe object
df = pd.DataFrame(pd.read_csv("completed_data.csv"))
  
# show the dataframe
df
```
