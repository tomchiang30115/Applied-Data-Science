```python
with open('barometer-last-year.csv', 'r') as file:
    data = []
    runtime = 0
    for line in file:
        variables = line.strip().split(',')
        if runtime == 0:
            pass
        else:
            data.append([variables[0], variables[1:]])
        runtime += 1
            
    for i in data:
        print(i[0], i[1])
```

```python
import pandas as pd

data2 = pd.DataFrame(data, columns=['DataFrame', 'Baro'])

print(data2)
```
